```python
import re
from collections import Counter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import urlextract
import emoji
from wordcloud import WordCloud
```

```python
# here i am basically generating the pandas dataframe form the chat-text file...... 
def generateDataFrame(file):
    with open(file, 'r', encoding='utf-8') as f:
        data = f.read()
    
    data = data.replace('\u202f', ' ')
    data = data.replace('\n', ' ')

    
    #this the regular expression pattern which i obtained from chatgpt by giving a chat sample......(it was really a tough task)
    
    dt_format = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:AM\s|PM\s|am\s|pm\s)?-\s'
    msgs = re.split(dt_format, data)[1:]
    date_times = re.findall(dt_format, data)
    date, time = [], []
    for dt in date_times:
        date.append(re.search(r'\d{1,2}/\d{1,2}/\d{2,4}', dt).group())
        time.append(re.search(r'\d{1,2}:\d{2}\s?(?:AM|PM|am|pm)?', dt).group())
    
    
    
    #now updating the table in better way to analyze the patterns........
    users, message = [], []
    for m in msgs:
        s = re.split(r'([\w\W]+?):\s', m)
        if len(s) < 3:
            users.append("Notifications")
            message.append(s[0])
        else:
            users.append(s[1])
            message.append(s[2])
    
    df = pd.DataFrame(list(zip(date, time, users, message)), columns=["Date", "Time(U)", "User", "Message"])
    return df
 
```

```python

#as i can see its working fine with a sample chat.....
generateDataFrame("b10chat.txt")
```

```python
# this is function to get unique users from the group or even two guys......
def getUsers(df):
    users = df['User'].unique().tolist()
    users.sort()
    users.remove('Notifications')
    users.insert(0, 'Everyone')
    return users
```

```python
#see its working here when you the pass the df which return by th fun generatedataframe....
getUsers(generateDataFrame("b10chat.txt"))
```

```python
# here i build this  Function to make the DataFrame more precise and preprocessed to analyze in detail
def PreProcess(df, dayfirst):
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=dayfirst)
    df['Time'] = pd.to_datetime(df['Time(U)']).dt.time
    df['year'] = df['Date'].apply(lambda x: int(str(x)[:4]))
    df['month'] = df['Date'].apply(lambda x: int(str(x)[5:7]))
    df['date'] = df['Date'].apply(lambda x: int(str(x)[8:10]))
    df['day'] = df['Date'].apply(lambda x: x.day_name())
    df['hour'] = df['Time'].apply(lambda x: int(str(x)[:2]))
    df['month_name'] = df['Date'].apply(lambda x: x.month_name())
    return df
```

```python


#checking the data after all......

PreProcess(generateDataFrame("b10chat.txt"), "mon")
```

```python
# this is Function to obtained the all   chat statistics includeing deleted messages,links,words,msg counts


def getStats(df):
    media = df[df['Message'] == "<Media omitted> "]
    media_cnt = media.shape[0]
    df.drop(media.index, inplace=True)
    
    deleted_msgs = df[df['Message'] == "This message was deleted "]
    deleted_msgs_cnt = deleted_msgs.shape[0]
    df.drop(deleted_msgs.index, inplace=True)
    
    temp = df[df['User'] == 'Notifications']
    df.drop(temp.index, inplace=True)
    
    extractor = urlextract.URLExtract()
    links = []
    for msg in df['Message']:
        x = extractor.find_urls(msg)
        if x:
            links.extend(x)
    links_cnt = len(links)
    
    word_list = []
    for msg in df['Message']:
        word_list.extend(msg.split())
    word_count = len(word_list)
    msg_count = df.shape[0]
    
    return df, media_cnt, deleted_msgs_cnt, links_cnt, word_count, msg_count
```

```python
#working fine..............

getStats(generateDataFrame("b10chat.txt"))
```
